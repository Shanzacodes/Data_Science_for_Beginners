```python
 #import libraries
 import seaborn as sns
 import matplotlib.pyplot as plt

 #load dataset
 flower = sns.load_dataset("iris")
 flower

 #draw a bar plot
 sns.barplot(x="petal_width", y="species", data=flower)
 plt.show()
```

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

#load dataset
boat = sns.load_dataset("titanic")
boat

#draw a bar plot
sns.barplot(x="fare", y="class", hue="who", data=boat, order=['Third','Second','First'], color= "pink")
plt.show()
```

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

#load dataset
boat = sns.load_dataset("titanic")
boat

#draw a bar plot
sns.barplot(x="fare", y="class", hue="who", data=boat, order=['Third','Second','First'], color= "pink", ci= None,
            palette="bright")
plt.show()
```

```python
#import libraries
import seaborn as sns
from numpy import mean
import matplotlib.pyplot as plt

#load dataset
boat = sns.load_dataset("titanic")
boat

#draw a bar plot
sns.barplot(x="fare", y="class", hue="who", data=boat, order=['Third','Second','First'], color= "pink", estimator=mean
            ,palette="bright")
plt.show()
```

```python
#import libraries
import seaborn as sns
from numpy import mean
import matplotlib.pyplot as plt

#load dataset
boat = sns.load_dataset("titanic")
boat

#draw a bar plot
sns.barplot(x="fare", y="class", hue="who", data=boat, order=['Third','Second','First'], color= "pink", estimator=mean
            ,saturation= 0.95,palette="bright")
plt.show()
```

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

#load dataset
boat = sns.load_dataset("titanic")
boat

#Draw a barplot
sns.barplot(x="class", y="fare", data=boat,
                linewidth=4,  facecolor=(1,1,1,0),
                errcolor='0',edgecolor='.5')
```
