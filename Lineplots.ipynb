**Adding** **titles**

```python
 #Import libraries
 import seaborn as sns
 import matplotlib.pyplot as plt

 #Load dataset
 pl = sns.load_dataset("iris")
 pl

 #Draw a line plot
 sns.lineplot(x='petal_length', y='petal_width', data=pl)
 plt.title("Petal length and width")
 
 plt.show()
```

**Adding** **Limits**

```python
 #Import libraries
 import seaborn as sns
 import matplotlib.pyplot as plt

 #Load dataset
 pl = sns.load_dataset("iris")
 pl

 #Draw a line plot
 sns.lineplot(x='petal_length', y='petal_width', data=pl)
 plt.title("Petal length and width")
 plt.xlim(1.5)
 plt.ylim(2)
 
 plt.show()
```

```python
 #Import libraries
 import seaborn as sns
 import matplotlib.pyplot as plt

 #Load dataset
 pl = sns.load_dataset("iris")
 pl

 #Adding style
 sns.set_style('whitegrid')
 #Draw a line plot
 sns.lineplot(x='petal_length', y='petal_width', data=pl)
 plt.title("Petal length and width")
 
 plt.show()
```

**Size** **of** **figure**

```python
 #Import libraries
 import seaborn as sns
 import matplotlib.pyplot as plt

 #Load dataset
 pl = sns.load_dataset("iris")

 #change figure
 plt.figure(figsize=(4,2))
 #Draw a line plot
 sns.lineplot(x='petal_length', y='petal_width', data=pl)
 plt.title("Petal length and width")
 
 plt.show()
```
